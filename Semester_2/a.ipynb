# K-NN Regression

```python
data = sns.load_dataset('diamonds')
data.head(2)
```

```python
y = data['price']
X = data.drop(['price'], axis = 1)
```

```python
cat = list(X.select_dtypes('category').columns)
num = list(X.select_dtypes('number').columns)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```

```python
model_scale = MinMaxScaler()
model_regressor = KNeighborsRegressor()
model_encoder = OneHotEncoder()
transformer = ColumnTransformer([('num', model_scale, num),
 ('cat', model_encoder, cat)])
model_combined = make_pipeline(transformer, model_regressor)
model_combined_optimal = GridSearchCV(model_combined,
param_grid = {'kneighborsregressor__n_neighbors': np.arange(3,4)},
 n_jobs = -1,
 return_train_score = True)
model_combined_optimal.fit(X_train, y_train);
```

```python
model_combined_optimal.best_params_
```

```python
model_combined_optimal.score(X_train,y_train)
```

```python
model_combined_optimal.score(X_test,y_test)
```

```python
k_grid = np.arange(1,71,2)
cv_scores = model_combined_optimal.cv_results_['mean_test_score']
```

```python
plt.figure(figsize = (5,5))
plt.plot(k_grid, cv_scores)
plt.show()
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2)
```
